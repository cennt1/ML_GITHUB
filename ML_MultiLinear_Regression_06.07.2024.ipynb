# Adjusted R2 Score (düzeltilmiş R2 degeri)

n = gözlem sayım (data noktası sayısı) 
p = feature

Adjusted R2 = 1 - [(1 - R2) * (n - 1) / (n - p - 1)]

```python
# n = gözlem sayım    # (data noktası sayısı) 
# p = feature

# Adjusted R2 = 1 - [(1 - R2) * (n - 1) / (n - p - 1)]
```

```python
def adj_r2(y_test, y_pred, df):
    r2 = r2_score(y_test, y_pred)         # R2 scor
    n = df.shape[0]                      # number of observations gözlem
    p =df.shape[1]-1                    # number of indep.variable = feature
    adj_r2 = 1 - (1-r2)*(n-1)/(n-p-1)   
    return adj_r2

#Bu kod, düzeltilmiş R^2 skorunu hesaplamak için bir fonksiyon tanımlar.
```

```python
adj_r2(y_test, y_pred, df)

#Bu kod, y_test, y_pred ve df değişkenlerini kullanarak düzeltilmiş R^2 skorunu hesaplar ve ekrana yazdırır.
```

```python
# Önceki r2 degerine yakın çıktı. çünkü zaten 3 feature vardı. 
# Aşağıdaki örnekte feature sayısını abartılı vererek adjusted etkisini görelim:
```

```python
def adj_r2(y_test, y_pred, df):
    r2 = r2_score(y_test, y_pred)
    n = df.shape[0]                             # number of observations(gözlem)
    p = 100                                    # number of independent variables ..feature sayısını 100 aldık.
    adj_r2 = 1 - (1-r2)*(n-1)/(n-p-1)
    return adj_r2
```

```python
adj_r2(y_test, y_pred, df)

# Bir görüşe göre, her eklenen feature için; 10-15 gözlem satırı daha eklenmelidir.
```

# Evaluation Model

```python
# Modelin genel degerlendirilmesini bu bölümde yapacağız.
```

```python
# Error metricleri hesaplayan ve yazan kod blogu

def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    print("Model testing performance: ")
    print("---------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```

```python
eval_metric(y_test, y_pred)
```

```python
# Overfitting kontrolü için train datasındaki R2 ve yukarıdaki test R2 (0.86) karşılaştırılır.
```
