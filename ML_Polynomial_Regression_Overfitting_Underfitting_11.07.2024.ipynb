```python
# Types of polinomials
   
# 1st degree ---> linear    b1x + b0   --->   1. derece

# 2nd degree ---> Quadratic b2x**2 + b1x + b0   --->    2.derece

# 3rd degree ---> Cubic     b3x**3 + b2x**2 + b1x + b0  --> 3.derece
```

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (10,6)
```

## Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
#örnek data seti üzerinden basitçe değerlendirelim.

data = np.array([[2,3,4]])
print(data)
```

```python
trans = PolynomialFeatures(degree=2, include_bias=False) 

# data'ya 2.dereceden Polynomial features ekleriz.
```

```python
trans.fit(data) 

# Dönüşümü 2 aşamada : fit ve transform olarak yaparız.
```

```python
trans.transform(data)    

# 2, 3, 4 , 2x3, 2x4, 3x4, 2**2, 3**2, 4**2 bu şekilde 2.dereceye çevirdi transform işlemi tamamladı.
# X1 ** 2 + X2 ** 2 + X3 ** 2 + X1X2 + X1X3 + X2X3 + X1 + X2 + X3 
```

```python
# hem transformer hem de fit işlemini aynı anda yapmak için alttaki kodu kullanabiliriz.

trans.fit_transform(data)
```

```python
# degree = 3 olsaydı;
# 2, 3, 4, 2x3, 2x4, 3x4, 2**2, 3**2, 4**2, 2x3x4, 3x2**2, 4x2**2, 
# 2x3**2, 4x3**2, 2x4**2, 3x4**2, 2**3, 3**3, 4**3
# şeklinde işlem yaparak features oluştururdu.
```

## Evaluation on the Test Set

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

# error metriclerini çağırıyoruz.
```

```python
y_pred = model_poly.predict(X_test)   
```

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    print("Model testing performance:")
    print("--------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```

```python
eval_metric(y_test, y_pred)  

# Test kısmının değerleri:
```

# Find an Optimum Polynomial Degree

```python
# RMSE değerine göre optimum dereceye karar vereceğiz,
# modeli daha da geliştirmeye çalışıyoruz.
```

```python
def poly(d):   # degree derecesi içine ne yazarsam ona göre çözüm oluşturacak
    
    train_rmse_errors = []   # trains rmse değerlerin toplayacak
    test_rmse_errors = []    # test rmse değerlerin toplayacak
    number_of_features = []  # derece arttıkça feature sayısının kaça çıktığnı göreceğiz
    
    for i in range(1, d):  # 1 den kaçıncı dereceye kadar istiyorsak ona göre yazarız
        polynomial_converter = PolynomialFeatures(degree = i, include_bias =False)
        poly_features = polynomial_converter.fit_transform(X)
        
        X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
        
        model = LinearRegression(fit_intercept=True) 
        model.fit(X_train, y_train)
        
        train_pred = model.predict(X_train)
        test_pred = model.predict(X_test)
        
        train_RMSE = np.sqrt(mean_squared_error(y_train,train_pred))
        test_RMSE = np.sqrt(mean_squared_error(y_test,test_pred))
        
        train_rmse_errors.append(train_RMSE)
        test_rmse_errors.append(test_RMSE)
        
        number_of_features.append(poly_features.shape[1])
        
    return pd.DataFrame({"train_rmse_errors": train_rmse_errors, "test_rmse_errors":test_rmse_errors, "number of features":number_of_features}, 
                        index=range(1,d))
```

```python
poly(10)

# Yukarıdaki fonk bize optimum polinom derecesindeki metricleri ve
# oluşan feature sayısını verir.
```

```python
# Yukarıdaki sonuçların grafiğini çizelim ve patlama noktasını görelim:

plt.plot(range(1,10), poly(10)["train_rmse_errors"], label = "TRAIN")
plt.plot(range(1,10), poly(10)["test_rmse_errors"], label = "TEST")
plt.xlabel("Polynamial Complex")
plt.ylabel("RMSE")
plt.legend();
plt.show()
```

```python
# grafiğe daha yakından bakalım:

plt.plot(range(1,6), poly(6)["train_rmse_errors"], label = "TRAIN")
plt.plot(range(1,6), poly(6)["test_rmse_errors"], label = "TEST")
plt.xlabel("Polynamial Complex")
plt.ylabel("RMSE")
plt.legend();
plt.show()
```

```python
# Patlama noktasının bir solundaki değerde kalmak daha güvenilir olacaktır.
# Bu nedenle 3.derece optimum deger olarak alınabilir.
```

# Finalize the model

```python
# optimum derecenin 3.derece oldugunu kabul ederek devam edelim, final modeli bu bilgiyle oluşturalım.
# modeli artık butun data ile eğiterek daha iyi sonuçlar elde etmek istiyoruz. Train-test kısmı artık yok.

final_poly_conv = PolynomialFeatures(degree=3, include_bias=False)

# 3.derece ile modelimizi eğiteceğiz.
```

```python
final_model = LinearRegression()
```

```python
final_model.fit(final_poly_conv.fit_transform(X), y)

# Modeli fit ederiz, X yerine Polymomial Feature kullanıp 3. derece için feature üretip
# Aldıgımız final_poly_conv değiskeni verip datanın hepsini (X) ile fit ederiz.
```

# Overfitting Uygulama Denemesi

```python
# Overfitting in yukarıdaki grafikte başladığı nokta olan (train ve test in ayrıştığı yer) 
# 5. dereceyi esas alarak feature  üretelim.
# Bilinçli şekilde yaptığımız overfitting oluşumunu izleyelim.
```

```python
over_poly_converter = PolynomialFeatures(degree=5, include_bias=False)

# 5.dereceden polynomial feature üretiyoruz.
```

```python
over_model =LinearRegression()   
```

```python
X_train, X_test, y_train, y_test = train_test_split(over_poly_converter.fit_transform(X), y, test_size=0.3, random_state=101)
```

```python
over_model.fit(X_train, y_train)
```

```python
y_pred_over = over_model.predict(X_test)
```

```python
eval_metric(y_test, y_pred_over)
```

```python
# train skorlarına bakalım

y_train_over = over_model.predict(X_train)
```

```python
eval_metric(y_train, y_train_over)
```

```python
# Train R2 si olan 0.99 ile test R2 si olan 0.76 arasında ciddi bir fark var. Bu fark bize overfitting i gösterdi.
# Train setin skorları çok iyi olmasına rağmen test setinin skorları kötü oldugu için overfitting teşhisi konuldu.
```
